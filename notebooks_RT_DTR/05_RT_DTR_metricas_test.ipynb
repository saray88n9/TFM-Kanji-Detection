# RT-DETR - Métricas del modelo final sobre test 

Se evalúan los modelos RT-DETR (original y denoised) sobre el split de test del dataset de parches, obteniendo las métricas oficiales de precisión, recall, mAP50 y mAP50-95.

## Metricas entrenamiento RT_DTR de parches en imagenes  original en test 1024

```python
from ultralytics import YOLO

model_parches_original_rtdetr = YOLO("/home/salvarado/TFM/resultados/result_RT_DTR/parches_original_1024_50ep_RT_DTR/weights/best.pt")
metrics_parches_original_rtdetr = model_parches_original_rtdetr.val(
    data="/home/salvarado/TFM/Dataset_preparados/dataset_kanji_parches/originales/dataset_parches_originales.yaml",
    
    split="test",
    imgsz=1024,
    device=2
)

print("=== RT-DETR Parches Original - Test ===")
print(f"Precisión: {metrics_parches_original_rtdetr.box.mp:.3f}")
print(f"Recall:    {metrics_parches_original_rtdetr.box.mr:.3f}")
print(f"mAP50:     {metrics_parches_original_rtdetr.box.map50:.3f}")
print(f"mAP50-95:  {metrics_parches_original_rtdetr.box.map:.3f}")
```

## Analisis

Precisión alta (0.831) pero recall bajo (0.747) → el modelo es bastante fiable cuando detecta algo, pero se le escapan muchos kanjis (falsos negativos).
mAP50-95 (0.526) notablemente más bajo que mAP50 (0.846) → cuando detecta correctamente, la caja no siempre queda bien ajustada al kanji real.

## Metricas entrenamiento RT_DTR de parches en imagenes  denoised en test 1024

```python
from ultralytics import YOLO

model_parches_denoised_rtdetr = YOLO("/home/salvarado/TFM/resultados/result_RT_DTR/parches_denoised_1024_50ep_RT_DTR/weights/best.pt")
metrics_parches_denoised_rtdetr = model_parches_denoised_rtdetr.val(
    data="/home/salvarado/TFM/Dataset_preparados/dataset_kanji_parches/denoised/dataset_parches_denoised.yaml",
    split="test",
    imgsz=1024,
    device=2
)

print("RT-DETR Parches Denoised - Test ")
print(f"Precisión: {metrics_parches_denoised_rtdetr.box.mp:.3f}")
print(f"Recall:    {metrics_parches_denoised_rtdetr.box.mr:.3f}")
print(f"mAP50:     {metrics_parches_denoised_rtdetr.box.map50:.3f}")
print(f"mAP50-95:  {metrics_parches_denoised_rtdetr.box.map:.3f}")
```

## Analisis

Mejora en todas las métricas respecto a Original, especialmente en recall (0.747→0.883) y mAP50 (0.846→0.905) → el modelo encuentra muchos más kanjis con la versión denoised.
A diferencia de YOLO (donde denoising no cambiaba nada), en RT-DETR sí parece haber una diferencia notable.
