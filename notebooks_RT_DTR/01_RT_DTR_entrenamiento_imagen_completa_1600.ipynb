### Entrenamiento - Imagen Completa Denoised

```python
contenido_yaml_denoised = f"""path: {ruta_denoised}
train: images/train
val: images/val
test: images/test

names:
  0: kanji
"""

with open(f"{ruta_denoised}/dataset_denoised.yaml", "w") as f:
    f.write(contenido_yaml_denoised)

print("Creado:", f"{ruta_denoised}/dataset_denoised.yaml")
```

```python
from ultralytics import YOLO

model_denoised = YOLO("rtdetr-l.pt")

results_denoised = model_denoised.train(
    data=f"{ruta_denoised}/dataset_denoised.yaml",
    epochs=50,
    imgsz=1600,
    batch=2,
    patience=10,
    project="/content/resultados_RT_DTR",
    name="imagen_completa_denoised_1600_50ep_RT_DTR",
    device=0,
    seed=42
)
```
