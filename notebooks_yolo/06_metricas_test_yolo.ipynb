# YOLO - Métricas de test (modelos finales con parches)

Se evalúan los modelos finales de YOLO, entrenados con parches a imgsz=1024, sobre el split de test del dataset correspondiente (original y denoised), para obtener las métricas 
oficiales de precisión, recall, mAP50 y mAP50-95.

## Parches Original  mejor modelo 1024

```python
from ultralytics import YOLO

model_original = YOLO("/home/salvarado/TFM/resultados/result_yolo/parches_originales_1024_50ep-2/weights/best.pt")
metrics_original = model_original.val(
    data="/home/salvarado/TFM/Dataset_preparados/dataset_kanji_parches/originales/dataset_parches_originales.yaml",
    split="test",
    imgsz=1024,
    device=1
)

print(" YOLO Parches Original - Test ")
print(f"Precisión: {metrics_original.box.mp:.3f}")
print(f"Recall:    {metrics_original.box.mr:.3f}")
print(f"mAP50:     {metrics_original.box.map50:.3f}")
print(f"mAP50-95:  {metrics_original.box.map:.3f}")
```

## Parches Denoised  mejor modelo 1024

```python
model_denoised = YOLO("/home/salvarado/TFM/resultados/result_yolo/parches_denoised_1024_50ep/weights/best.pt")
metrics_denoised = model_denoised.val(
    data="/home/salvarado/TFM/Dataset_preparados/dataset_kanji_parches/denoised/dataset_parches_denoised.yaml",
    split="test",
    imgsz=1024,
    device=1
)

print(" YOLO Parches Denoised - Test ")
print(f"Precisión: {metrics_denoised.box.mp:.3f}")
print(f"Recall:    {metrics_denoised.box.mr:.3f}")
print(f"mAP50:     {metrics_denoised.box.map50:.3f}")
print(f"mAP50-95:  {metrics_denoised.box.map:.3f}")
```
